```python
import pandas as pd
import numpy as np
```

```python
#Task 1 : Create employee,project,and seniority dataframes
employee = pd.DataFrame({'ID':['A001','A002','A003','A004','A005'],
                         'Name':['John Cruise','Alice Luxemburg','Tom Cruise','Nina Agara','Amy Johny'],
                         'Gender':['M','F','M','F','F'],
                         'City':['Paris','London','Berlin','Newyork','Madrid'],
                         'Age':[25,27,29,31,30]})
project = pd.DataFrame({'ID':['A001','A002','A003','A004','A005','A002','A005','A003','A001','A003','A001','A004','A004','A005'],
                        'Project':['Project 1','Project 2','Project 3','Project 4','Project 5','Project 6','Project 7','Project 8','Project 9',
                                   'Project 10','Project 11','Project 12','Project 13','Project 14'],
                        'Cost':[1002000,2000000,4500000,5500000,np.nan,68000,400000,350000,300000,2000000,np.nan,3000000,3000000,2000000],
                        'Status':['Finished','Ongoing','Finished','Ongoing','Finished','Failed','Finished','Failed','Ongoing','Finished','Failed',
                                  'Ongoing','Finished','Finished']})
seniority = pd.DataFrame({'ID':['A001','A002','A003','A004','A005'],
                          'Designation_level':[2,2,3,2,3]})

# Save the datsaframes as CSV files
employee.to_csv("employee.csv",index=False)
project.to_csv("project.csv",index=False)
seniority.to_csv("seniority.csv",index=False)
```

```python
employee_df=pd.read_csv("employee.csv")
project_df=pd.read_csv("project.csv")
seniority_df=pd.read_csv("seniority.csv")

employee_df.head()
```

```python
# Task 2:(Replace missing cost values with running average)
# Initialize variables to keep track of total cost and count
running_total=0
count=0

# Loop through each row in the dataframe
for i in range(len(project_df)):

#Get the current cost value
    cost=project_df.loc[i,'Cost']

# Check if the cost is not missing
    if pd.notna(cost):
        running_total+=cost
        count+=1
# if the value is missing replace it with running average
    else:
         project_df.loc[i,'Cost']=running_total/count

# Display numbers without scientific notation
pd.options.display.float_format='{:.0f}'.format
print(project_df)
```

```python
#Task 3: 
#Split 'Name'column into 'first_name' and 'last_name'
employee_df[['First_name','Last_name']]=employee_df['Name'].str.split(' ',expand=True)
# Remove the original 'Name' column
employee_df.drop('Name',axis=1,inplace=True)

employee_df
```

```python
#Task 4:
# Merge employee,seniority,and project dataframes 
# Using 'ID' as the common key
Final=employee_df.merge(seniority_df,on='ID').merge(project_df,on='ID')

print(Final)
```

```python
#Task 5
# Assign 5% bonus only if project status is 'Finished'
# np.where condition applies bonus
Final['Bonus']=np.where(Final['Status']=='Finished',Final['Cost']*0.05,0)

print(Final[['ID','Project','Status','Bonus']])
```

```python
# Task 6(Demote designation level by 1 for failed projects)
# Demote(increase level number)
Final.loc[Final['Status']=='Failed','Designation_level']+=1
# Remove employees with level > 4
Final=Final[Final['Designation_level']<=4]

print(Final[['ID','Project','Status','Designation_level']])
```

```python
#Task 7:
# Add title(Mr./Mrs.) to first_name based on gender
Final['First_name']=np.where(Final['Gender']=='M',
                          'Mr. '+Final['First_name'],
                          'Mrs. '+ Final['First_name'])
# And drop gender column
Final.drop('Gender',axis=1,inplace=True)
print(Final[['ID','First_name']])
```

```python
# Task 8(Promote designation level by 1 if age>29)
for i in range(len(Final)):
# Check if age is greater than 29
    if Final.loc[i,'Age']>29:
# Promote( decrease designation level since 1 is the highest)        
        Final.loc[i,'Designation_level']-=1
# Ensure designation level does not go below 1        
Final['Designation_level']=Final['Designation_level'].clip(lower=1)

print(Final[['ID','Age','Designation_level']])
```

```python
#Task 9:
# Calculate total project cost for each employee using groupby
Total_project_cost=Final.groupby(['ID','First_name']) ['Cost'].sum().reset_index()
# Renaming columns for better readability
Total_project_cost.rename(columns={'Cost' :'Totalcost'},inplace=True)

print(Total_project_cost)
```

```python
#Task 10 
# Filter employees whose city contains the letter 'o'
o_city=Final[Final['City'].str.contains('o', case=False)]

print(o_city[['ID','First_name','Last_name','City','Age']])
```
